# Contrast Adjustment

**Contrast adjustment is a technique used to improve the visibility of features in an image by enhancing the difference between the light and dark areas**.

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("ORGINAL IMAGE")
cv2_imshow(image)
image_rbg=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
print("RBG")
cv2_imshow(image_rbg)
alpha=1.5
beta=50
adjusted=cv2.convertScaleAbs(image_rbg,alpha=alpha,beta=beta)
print("ADJUSTED CONTRAST IMAGE")
cv2_imshow(adjusted)
```

# HISTOGRAM EQUILIZER

**Histogram equalization is a method used to enhance the contrast of an image by transforming its intensity values so that the histogram of the output image is evenly distributed**

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("ORGINAL IMAGE")
cv2_imshow(gray_image)
eq=cv2.equalizeHist(gray_image)
print("EQUALIZED IMAGE")
cv2_imshow(eq)
```

# NOISE REDUCTION

**Noise reduction techniques are used to remove unwanted random variations in brightness or color, known as noise, from an image.**

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("ORGINAL IMAGE")
cv2_imshow(image)
print("BLURRED IMAGE")
blurred=cv2.medianBlur(image,3)
cv2_imshow(blurred)
print("BILATERAL IMAGE")
bilateral=cv2.bilateralFilter(image,9,75,75)
cv2_imshow(bilateral)
print("GAUSSIAN IMAGE")
gaussian=cv2.GaussianBlur(image,(5,5),0)
cv2_imshow(gaussian)
```

# INPAINTING THE IMAGE

**Inpainting involves reconstructing lost or deteriorated parts of an image. This technique is often used for restoring old photographs, removing objects, or filling in missing data. Algorithms for inpainting include patch-based methods and partial differential equations (PDE) based methods.**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("ORGINAL IMAGE")
cv2_imshow(image)

mask = np.zeros(image.shape[:2], dtype="uint8")

mask[100:400, 100:400] = 255
print("MASKING IMAGE")
cv2_imshow(mask)
inpaint = cv2.inpaint(image, mask, 3, cv2.INPAINT_NS)
print("INPAINTED IMAGE")
cv2_imshow(inpaint)
```

# DEBLURRING THE IMAGE

**Deblurring techniques are used to restore sharpness to an image that has been blurred due to factors like camera shake or motion.**

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("ORGINAL IMAGE")
cv2_imshow(image)
print("BLURRED IMAGE")
blurred=cv2.medianBlur(image,3)
cv2_imshow(blurred)
print("BILATERAL IMAGE")
bilateral=cv2.bilateralFilter(image,9,75,75)
cv2_imshow(bilateral)
print("GAUSSIAN IMAGE")
gaussian=cv2.GaussianBlur(image,(5,5),0)
cv2_imshow(gaussian)
unsharp=cv2.addWeighted(image,1.5,gaussian,-0.5,0)
print("UNSHARP IMAGE")
cv2_imshow(unsharp)
```

# LINEAR FILTERING


**Linear filtering is a process in which the value of the output pixel is linear combinations of the neighboring input pixels. This process is done by a technique called **Convolution**.

**Convolution is the process of adding each element of the image to its local neighbors, weighted by the kernel.****

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/images.jpeg")
print("ORGINAL IMAGE")
cv2_imshow(image)
kernel=np.ones((5,5),dtype="float")/25
print("KERNEL")
print(kernel)
print("FILTERED IMAGE")
filtered=cv2.filter2D(image,-1,kernel)
cv2_imshow(filtered)
```
