##  INDEPENDENT COMPONENT ANALYSIS

**Independent Component Analysis or short for ICA is a technique for separating a multivariate signal into its underlying component. ICA helps in the extraction of the desired component from the mixture of multiple components or signals.**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from sklearn.decomposition import FastICA
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/download.png")
print("ORGINAL IMAGE")
cv2_imshow(image)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("GRAY IMAGE")
cv2_imshow(gray)
print("RESHAPED IMAGE")
gray=gray.reshape((gray.shape[0]*gray.shape[1],1))
print(gray.shape)
ica=FastICA(n_components=1)
components=ica.fit_transform(gray)
print(components.shape)
print("COMPONENTS")
print(components)



```

# EDGE DETECTION

**Edge detection involves identifying the boundaries within an image. Techniques like the Sobel, Canny, and Prewitt operators are used to detect edges by finding areas of high intensity gradient.**

*SOBEL EDGE DETECTION*

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("ORGINAL IMAGE")
cv2_imshow(image)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("GRAY IMAGE")
cv2_imshow(gray)
sobelx=cv2.Sobel(gray,cv2.CV_64F,1,0)
sobely=cv2.Sobel(gray,cv2.CV_64F,0,1)
sobelxy=cv2.Sobel(gray,cv2.CV_64F,1,1)
print("SOBELX IMAGE")
cv2_imshow(sobelx)
print("SOBELY IMAGE")
cv2_imshow(sobely)
print("SOBELXY IMAGE")
cv2_imshow(sobelxy)
```

*CANNY EDGE DETECTION*

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/images.jpeg")
print("ORGINAL IMAGE")
cv2_imshow(image)
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("GRAY IMAGE")
blurred=cv2.GaussianBlur(gray_image,(5,5),0)
print("BLURRED IMAGE")
cv2_imshow(blurred)
canny=cv2.Canny(blurred,30,150)
print("CANNY IMAGE")
cv2_imshow(canny)
```

*LAPLACIAN EDGE DETECTION*

```python

import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/images.jpeg")
print("ORGINAL IMAGE")
cv2_imshow(image)
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("GRAY IMAGE")
blurred=cv2.GaussianBlur(gray_image,(5,5),0)
print("BLURRED IMAGE")
cv2_imshow(blurred)
laplacian=cv2.Laplacian(blurred,cv2.CV_64F)
print("LAPLACIAN IMAGE")
cv2_imshow(laplacian)
```
