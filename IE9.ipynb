# *TEXTURE SYNTHESIS*

**Texture synthesis generates large textures from small sample images, ensuring that the generated texture looks natural and continuous. This technique is widely used in computer graphics and game design.**

PATCH BASED SYNTHESIS

**This method generates a new texture by randomly selecting patches from the sample texture and placing them into the output texture while maintaining the overall structure and coherence.**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/dog.jpeg")
print("ORGINAL IMAGE")
cv2_imshow(image)
rows,cols,channels=image.shape
output=np.zeros((rows,cols,channels),dtype=np.uint8)
for i in range(1000):
  row=np.random.randint(0,rows)
  col=np.random.randint(0,cols)
  patch=image[row:row+100,col:col+100]
  output[row:row+100,col:col+100]=patch
print("PATCH BASED SYNTHESIS IMAGE")
cv2_imshow(output)

```

*Pixel-based Texture Synthesis*

**This method synthesizes textures one pixel at a time by filling each pixel with values that match the statistics (e.g., color distribution) of the input texture.**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/dog.jpeg")
print("ORGINAL IMAGE")
cv2_imshow(image)
rows,cols,channels=image.shape
output=np.zeros((rows,cols,channels),dtype=np.uint8)
for i in range(rows):
  for j in range(cols):
    output[i,j]=image[np.random.randint(0,rows),np.random.randint(0,cols)]
print("PIXEL BASED SYNTHESIS IMAGE")
cv2_imshow(output)


```

*Neural Network-based Texture Synthesis*

**Using deep learning models like GANs (Generative Adversarial Networks) to generate highly realistic textures.**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/dog.jpeg")
print("ORGINAL IMAGE")
cv2_imshow(image)
rows,cols,channels=image.shape
output=np.zeros((rows,cols,channels),dtype=np.uint8)
for i in range(rows):
  for j in range(cols):
    output[i,j]=image[np.random.randint(0,rows),np.random.randint(0,cols)]
    output[i,j]=image[np.random.randint(0,rows),np.random.randint(0,cols)]
print("NEURAL NETWORK BASED SYNTHESIS IMAGE")
cv2_imshow(output)
```
