# CONVERTING RBG COMPONENT TO CMY COMPONENT

```python
import cv2
from google.colab.patches import cv2_imshow
print("ORIGINAL IMAGE")
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/god.jpeg")
cv2_imshow(image)
r,g,b=cv2.split(image)
print("R Component")
cv2_imshow(r)
print("G Component")
cv2_imshow(g)
print("B Component")
cv2_imshow(b)
print("C Component")
c=1-r
cv2_imshow(c)
print("M Component")
m=1-g
cv2_imshow(m)
print("Y Component")
y=1-b
cv2_imshow(y)
```

# IMUTILS LIBRARY

```python
import cv2
import imutils
print("ORIGINAL IMAGE")
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/god.jpeg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print("ROTATED IMAGE")
r1=imutils.rotate(image,45)
cv2_imshow(r1)
print("TRANSLATIONAL IMAGE")
R2=imutils.translate(image,25,75)
cv2_imshow(R2)
print("RESIZED IMAGE")
r3=imutils.resize(image,width=125)
cv2_imshow(r3)
print("RESIZE IMAGE WITHOUT ASPECT RATIO")
r4=cv2.resize(image,(125,125))
cv2_imshow(r4)

```
